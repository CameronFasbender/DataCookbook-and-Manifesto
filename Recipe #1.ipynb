# Recipe #1: Making a new column in a dataframe by applying a function

```python
df = pd.read_csv("space_missions.csv", encoding = "latin1")
df
```

```python
def time_of_day(military_time):
    hour = int(military_time[:2])
    if hour >= 20 or hour < 4:
        return 'Night'
    elif hour >= 4 and hour < 13:
        return 'Morning'
    else:
        return 'Afternoon'
    
#Function that extracts the hour values from the Time column and assigns them a string description based on their value.
```

```python
df['Time'] = df['Time'].fillna('00:00') 

#Converts all empty Time rows to 00:00 so that the function can be applied to the entire dataframe.
```

```python
df['Time of Day'] = df['Time'].apply(time_of_day)
df

#Applies the time_of_day funtion to the Time columns.
#Saves results in a new column, Time of Day.
```
