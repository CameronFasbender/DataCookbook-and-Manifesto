```python
import pandas as pd
import numpy as np
import json
```

# Recipe #2: Converting a json file into a csv, then saving.

```python
with open('StreamingHistory0.json', 'r', encoding='utf-8') as f:
    data = json.load(f)
#"with" ensures that the file is closed properly after the contents inside the open() statement are completed
#'r' specifies that the file is being opened for read only
    
df = pd.json_normalize(data)
#normalization makes each key from the json file into a column for tabular formatting

df.to_csv('StreamingHistory0.csv', index=False)
df
#converting json to csv
```

```python
file_path = ("~/Downloads/StreamingHistoy0.csv")
df.to_csv(file_path, index=False)

#the to_csv method is able to take a file path as an argument, this will save the csv to that filepath on your computer
#index=False is so that the row index is not its own column making the dataframe look cleaner.
```
