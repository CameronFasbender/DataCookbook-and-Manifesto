# Recipe 3: Merging dataframes

```python
df1 = pd.read_csv("StreamingHistory5.csv")
df2 = pd.read_csv("StreamingHistory0.csv")
```

```python
df1
```

```python
df2
```

```python
concat_df = pd.concat([df1, df2], axis=0)
concat_df

#concat is used to combine dataframes vertically, essntially stacking all of one on top of the other
#this dataframe now has the 555 additional songs from the first dataframe on top of the second
```

```python
df_merged = pd.merge(df1, df2, on=['artistName', 'trackName'], how='outer')
#merges the dataframes on the artistName and trackName columns
#outer merge merges all rows from both data frames and fills in missing values with NaN

df_merged.drop_duplicates(subset=['artistName', 'trackName'], keep='first', inplace=True)
#because some songs were played more than once in the first dataframe, it is important to only keep one unique identifier from it.
#this is done by using the drop duplicates function
#keep='first' means that the first instance of a duplicate is the one that will be kept

pd.set_option('display.max_rows', None)
#allows for the entire table to be viewed

df_merged
```
